```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(16,8))
den = sch.dendrogram(sch.linkage(df,method='ward'))
plt.title('dendrogram')
plt.xlabel('Houses')
plt.ylabel('Euclident Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5,affinity = 'euclidean' , linkage = 'average')
y_pred = hc.fit_predict(df)
y_pred
```

```python
X = df.iloc[ : , [0,1]].values
```

```python
X
```

```python
plt.figure(figsize=(15,8))
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],color='red',label='cluster 1',s=50)
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],color='blue',label='cluster 1',s=50)
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],color='grey',label='cluster 1',s=50)
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],color='black',label='cluster 1',s=50)
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],color='yellow',label='cluster 1',s=50)
plt.title('Hierarchy clustering')
plt.xlabel('MedInc')
plt.ylabel('Hou_Age')
plt.show()
```
